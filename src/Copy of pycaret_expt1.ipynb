```python
import dvc.api
```

```python
with dvc.api.open("data/creditcard.csv", repo = "https://github.com/sanidhyaanand/MLOps_Assignment") as fd:
    df = pd.read_csv(fd)

df.head()
```

```python
y = df.Class
X = df.drop(["Class", "Time"], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1,random_state=42,stratify=y)

train_df = pd.concat([X_train, y_train], axis=1)
test_df = pd.concat([X_test, y_test], axis=1)

train_df.to_csv("./data/processed/train.csv")
test_df.to_csv("./data/processed/test.csv")
```

```python
X_train, X_test, y_train, y_test = train_test_split(test_df.drop(["Class"], axis = 1),test_df.Class,test_size=0.1,random_state=42,stratify=test_df.Class)

train_df = pd.concat([X_train, y_train], axis=1)
test_df = pd.concat([X_test, y_test], axis=1)
```

```python
print("train size is: ", train_df.size)
print("test size is: ", test_df.size)
```

```python
from pycaret.classification import *
```

```python
expt_cfn = setup(data = train_df,
                 target = "Class",
                 session_id = 42,
                 imputation_type = "iterative",
                 categorical_imputation="mode",
                 fix_imbalance = True)
```

```python
models()
```

```python
best_clf = compare_models(sort="F1", fold=5, exclude = ["lr", "ada"])
```

```python
et_clf = create_model('et', fold = 5)
lgbm_clf = create_model('lightgbm', fold = 5)
gb_clf = create_model('gbc', fold = 5)
knn_clf = create_model('knn', fold = 5)
```

```python
tuned_et_clf = tune_model(et_clf, fold = 5, optimize = "F1", search_algorithm="random")
tuned_lgbm_clf = tune_model(lgbm_clf, fold = 5, optimize = "F1", search_algorithm="random")
tuned_gb_clf = tune_model(gb_clf, fold = 5, optimize = "F1", search_algorithm="random")
# tuned_ada_clf = tune_model(ada_clf, fold = 5, optimize = "F1", search_algorithm="random")
tuned_knn_clf = tune_model(knn_clf, fold = 5,optimize = "F1", search_algorithm="random")
```

```python
rf_clf = create_model('rf', fold = 5)
```

```python
tuned_rf_clf = tune_model(rf_clf, fold = 5, optimize = "F1", search_algorithm="random")
```

```python
best_tuned = compare_models(sort="F1", fold=5, include = [tuned_et_clf, tuned_rf_clf, tuned_gb_clf, tuned_knn_clf, tuned_lgbm_clf])
```

```python
blended_model = blend_models(estimator_list=[tuned_lgbm_clf, tuned_knn_clf, tuned_et_clf, tuned_gb_clf, tuned_rf_clf])
```

```python
predict_model(blended_model)
```

```python
plot_model(blended_model, plot = 'confusion_matrix')
```

```python
unseen_predictions = predict_model(blended_model, data = test_df)
unseen_predictions.head()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("Accuracy:\t", accuracy_score(unseen_predictions.Class, unseen_predictions.Label))
print("Precision Score:\t", precision_score(unseen_predictions.Class, unseen_predictions.Label))
print("Recall Score:\t", recall_score(unseen_predictions.Class, unseen_predictions.Label))
print("F1 Score:\t", f1_score(unseen_predictions.Class, unseen_predictions.Label))
```
